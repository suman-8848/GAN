```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras import layers, models, Model
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout,LeakyReLU, Reshape, BatchNormalization

tf.random.set_seed(42)
np.random.seed(42)

# Load and preprocess data
(x_train, _), (_, _) = fashion_mnist.load_data()

# Normalize and reshape
x_train = x_train / 127.5 - 1.0
x_train = x_train.reshape(-1, 28, 28, 1).astype(np.float32)
```

```python

def build_generator(latent_dim=100):
    model = models.Sequential([
        # Input layer
        Dense(7*7*256, input_dim=latent_dim, use_bias=False),
        BatchNormalization(),
        LeakyReLU(alpha=0.2),

        # Reshape for transposed convolutions
        Reshape((7, 7, 256)),

        # Upsampling
        layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False),
        BatchNormalization(),
        LeakyReLU(alpha=0.2),

        layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False),
        BatchNormalization(),
        LeakyReLU(alpha=0.2),

        # Output layer
        layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', activation='tanh')
    ])
    return model

def build_discriminator(input_shape=(28, 28, 1)):

    model = models.Sequential([
        # First convolutional layer
        Conv2D(64, (5, 5), strides=(2, 2), padding='same', input_shape=input_shape),
        LeakyReLU(alpha=0.2),
        Dropout(0.3),

        # Second convolutional layer
        Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        LeakyReLU(alpha=0.2),
        Dropout(0.3),

        # Flatten and output
        Flatten(),
        Dense(1, activation='sigmoid')
    ])
    return model
```

```python

class FashionMNISTGAN(Model):
    def __init__(self, generator, discriminator, latent_dim=100):
        super(FashionMNISTGAN, self).__init__()
        self.generator = generator
        self.discriminator = discriminator
        self.latent_dim = latent_dim

    def compile(self,gen_optimizer,disc_optimizer,gen_loss_fn,disc_loss_fn):
        super(FashionMNISTGAN, self).compile()
        self.gen_optimizer = gen_optimizer
        self.disc_optimizer = disc_optimizer
        self.gen_loss_fn = gen_loss_fn
        self.disc_loss_fn = disc_loss_fn

    def train_step(self, real_images):
        # Batch size from input images
        batch_size = tf.shape(real_images)[0]

        # Generate noise
        noise = tf.random.normal([batch_size, self.latent_dim])

        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            # Generate fake images
            generated_images = self.generator(noise, training=True)

            # Discriminator predictions
            real_output = self.discriminator(real_images, training=True)
            fake_output = self.discriminator(generated_images, training=True)


            # fooling the discriminator
            gen_loss = self.gen_loss_fn(tf.ones_like(fake_output), fake_output)

            disc_loss_real = self.disc_loss_fn(tf.ones_like(real_output), real_output)
            disc_loss_fake = self.disc_loss_fn(tf.zeros_like(fake_output), fake_output)
            disc_loss = 0.5 * (disc_loss_real + disc_loss_fake)

        # Calculate gradients
        gen_gradients = gen_tape.gradient(gen_loss, self.generator.trainable_variables)
        disc_gradients = disc_tape.gradient(disc_loss, self.discriminator.trainable_variables)

        # Apply gradients with gradient
        self.gen_optimizer.apply_gradients(zip(gen_gradients, self.generator.trainable_variables))
        self.disc_optimizer.apply_gradients(zip(disc_gradients, self.discriminator.trainable_variables))

        return {"disc_loss": disc_loss, "gen_loss": gen_loss}

def generate_and_save_images(model, epoch, test_input): #saving generated image

    generated_images = model.generator(test_input, training=False)
    generated_images = (generated_images * 127.5 + 127.5).numpy().reshape((-1, 28, 28))

    plt.figure(figsize=(10, 10))
    for i in range(25):
        plt.subplot(5, 5, i+1)
        plt.imshow(generated_images[i], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig(f'generated_images_epoch_{epoch}.png')
    plt.close()
```

```python


EPOCHS = 50
BATCH_SIZE = 64
LATENT_DIM = 100

train_dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(10000).batch(BATCH_SIZE)

# Initialize models
generator = build_generator(LATENT_DIM)
discriminator = build_discriminator()
gen_optimizer = Adam(learning_rate=0.0001, beta_1=0.5)
disc_optimizer = Adam(learning_rate=0.0002, beta_1=0.5)
gen_loss_fn = BinaryCrossentropy()
disc_loss_fn = BinaryCrossentropy()

# Create GAN model
fashion_gan = FashionMNISTGAN(generator, discriminator, LATENT_DIM)
fashion_gan.compile(
    gen_optimizer=gen_optimizer,
    disc_optimizer=disc_optimizer,
    gen_loss_fn=gen_loss_fn,
    disc_loss_fn=disc_loss_fn
)

test_noise = tf.random.normal([25, LATENT_DIM])
#  visualization
loss_history = {'gen_loss': [], 'disc_loss': []}

for epoch in range(EPOCHS):
    for batch in train_dataset:
        result = fashion_gan.train_step(batch)

    # Store loss history
    loss_history['gen_loss'].append(result['gen_loss'].numpy())
    loss_history['disc_loss'].append(result['disc_loss'].numpy())

    print(f"Epoch {epoch+1}: D Loss = {result['disc_loss']:.4f}, G Loss = {result['gen_loss']:.4f}")

    # Generate and save images every 10 epochs
    if (epoch + 1) % 10 == 0:
        generate_and_save_images(fashion_gan, epoch + 1, test_noise)

# Plot loss history
plt.figure(figsize=(10,5))
plt.plot(loss_history['gen_loss'], label='Generator Loss')
plt.plot(loss_history['disc_loss'], label='Discriminator Loss')
plt.title('GAN Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.savefig('gan_loss_plot.png')
plt.close()

# Final image generation
generate_and_save_images(fashion_gan, EPOCHS, test_noise)
```
